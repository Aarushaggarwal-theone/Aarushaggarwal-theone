```python
cols=[
 'FGM_2',
 'FGA_2',
 'FGA_3',
 'FGM_3',
 'FTA',
 'FTM',
 'AST',
 'BLK',
 'STL',
 'DREB',
 'OREB',
 'F_tech','TOV','TOV_team',
 'F_personal',
 'largest_lead','team_compared_to_comp','team_score','possession','target']


new_cols=[f"{c}_rolling" for c in cols]
```

```python
def rolling_averages(group, cols, new_cols):
  group=group.sort_values('game_date')
  rolling_stats=group[cols].rolling(3, closed='left').mean()
  group[new_cols]=rolling_stats
  group=group.dropna(subset=new_cols)
  return group
```

```python
match_rolling=match.groupby('team').apply(lambda x: rolling_averages(x, cols, new_cols))
match_rolling
```
